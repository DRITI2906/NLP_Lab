```python
import json
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
import os
import urllib.request
```

```python
font_path = "NotoSansGujarati-Regular.ttf"  # Already downloaded manually
font_prop = fm.FontProperties(fname=font_path)
```

```python
# Load the font
font_prop = fm.FontProperties(fname=font_path)
plt.rcParams['font.family'] = font_prop.get_name()
plt.rcParams['font.sans-serif'] = [font_prop.get_name()]
plt.rcParams['axes.unicode_minus'] = False
```

```python
with open("tokenized.json", "r", encoding="utf-8") as f:
    data = json.load(f)
```

```python
all_tokens = [token for sentence in data["tokens"] for token in sentence]
```

```python
freq_dist = {}
for token in all_tokens:
    freq_dist[token] = freq_dist.get(token, 0) + 1
```

```python
sorted_freq = sorted(freq_dist.items(), key=lambda x: x[1], reverse=True)
```

```python
top_100 = sorted_freq[:100]
words, freqs = zip(*top_100)
```

```python
plt.figure(figsize=(15,6))
plt.bar(words, freqs)
plt.xticks(rotation=90, fontproperties=font_prop)
plt.title("Top 100 Most Frequent Gujarati Words (Before Stopword Removal)", fontproperties=font_prop)
plt.xlabel("શબ્દો", fontproperties=font_prop)
plt.ylabel("આવર્તન", fontproperties=font_prop)
plt.show()
```

```python
# Removing the Stopwords
for threshold in [10, 20, 30]:
    stop_words = [word for word, freq in freq_dist.items() if freq > threshold]

    filtered_tokens = [token for token in all_tokens if token not in stop_words]

    filtered_freq = {}
    for token in filtered_tokens:
        filtered_freq[token] = filtered_freq.get(token, 0) + 1

    sorted_filtered = sorted(filtered_freq.items(), key=lambda x: x[1], reverse=True)[:100]
    words, freqs = zip(*sorted_filtered)
```

```python
plt.figure(figsize=(15,6))
plt.bar(words, freqs)
plt.xticks(rotation=90, fontproperties=font_prop)
plt.title(f"Top 100 Gujarati Words After Stopword Removal (Threshold = {threshold})", fontproperties=font_prop)
plt.xlabel("શબ્દો", fontproperties=font_prop)
plt.ylabel("આવર્તન", fontproperties=font_prop)
plt.show()
```
